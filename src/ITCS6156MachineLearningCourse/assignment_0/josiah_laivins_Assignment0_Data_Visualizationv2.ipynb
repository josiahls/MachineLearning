# Data - Regression

The link this data set can be found [at this Kaggle climate change link](https://www.kaggle.com/berkeleyearth/climate-change-earth-surface-temperature-data/home)

Climate change seemed appropriate for regression. 
The main set I was interested in was the Global Temperatures. The data set also has temperatures filtered by City, Country and State. 
The data ranges from years 1750 to around 2015. 

The features in the GlobalTemperatures csv are: 
- 'datetime'
- 'LandAverageTemperature'
- 'LandAverageTemperatureUncertainty'
- 'LandMaxTemperature'
- 'LandMaxTemperatureUncertainty',
- 'LandMinTemperature'
- 'LandMinTemperatureUncertainty',
- 'LandAndOceanAverageTemperature',
- 'LandAndOceanAverageTemperatureUncertainty'

There are 3179 records in this csv, however there are other csv's that have records up to 577462 records such as the GlobalLandTemperaturesByCountry csv.

```python
from pathlib import Path
```

```python
import datetime as dt

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

nrows = 200

# Get Data
data = pd.read_csv(str(Path().absolute()) + '/data/climate-change-earth-surface-temperature-data/GlobalTemperatures.csv', nrows=nrows)

data.dropna(subset=['LandAverageTemperature'], inplace=True)
data = data.reset_index(drop=True)
past_dt = list(pd.to_datetime(data['dt']))
xaxis = past_dt
nrows = data.shape[0]

# Set the target line
target = np.array(data['LandAverageTemperature']).reshape(-1, 1)

# Build plot
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(target, label='Recorded Temperatures')

# Make it look nice
fig.autofmt_xdate()
dates = [d.strftime('%Y/%m/%d') for d in xaxis]
plt.xticks(range(0, len(dates), int(len(dates) / 20+2)), dates[::int(len(dates) / 20+2)])
plt.title(f'Prediction using: {nrows} samples')
plt.ylabel('Average Land Temperature')
plt.xlabel('Time Period (YYYY/MM/DD)')
plt.legend()
plt.show()
```

```python
nrows = None

# Get Data
data = pd.read_csv('./data/climate-change-earth-surface-temperature-data/GlobalLandTemperaturesByCountry.csv', nrows=nrows)
data = data[data['Country'] == 'United States']

data.dropna(subset=['AverageTemperature'], inplace=True)
data = data.reset_index(drop=True)
past_dt = list(pd.to_datetime(data['dt']))
xaxis = past_dt
nrows = data.shape[0]

# Set the target line
target = np.array(data['AverageTemperature']).reshape(-1, 1)

# Build plot
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(target, label='Recorded Temperatures')

# Make it look nice
fig.autofmt_xdate()
dates = [d.strftime('%Y/%m/%d') for d in xaxis]
plt.xticks(range(0, len(dates), int(len(dates) / 20+2)), dates[::int(len(dates) / 20+2)])
plt.title(f'Prediction using: {nrows} samples for the United States')
plt.ylabel('Average Temperature')
plt.xlabel('Time Period (YYYY/MM/DD)')
plt.legend()
plt.show()
```

The link this data set can be found [this Kaggle Stack Overflow link](https://www.kaggle.com/stackoverflow/stack-overflow-2018-developer-survey)

The data set found from stack overflow seemed interesting to use for classification. There should be interesting insights that I would expect to see. For example is the Undergrad Major correlated with Job Satisfaction? Does WakeTime affect someone's eating habits?

There are 98855 records found in this data set. There are 129 features, but I am going to list some main ones that I was immediately interested in:

- 'Student'

- 'YearsCoding'
- 'JobSatisfaction'
- 'CareerSatisfaction'
- 'SkipMeals'
- 'WakeTime'
- 'HoursComputer'
- 'RaceEthnicity' 


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
%matplotlib inline

nrows = None
# get_linear_classification(nrows)

# Get Data
data = pd.read_csv('./data/stack-overflow-2018-developer-survey/survey_results_public.csv', nrows=nrows)
# data.dropna(subset=['FormalEducation', 'JobSatisfaction'], inplace=True)
data = data.reset_index(drop=True)
nrows = data.shape[0]

# Convert String columns into one-hot encoded columns
for column in data:
    if data[column].dtype == object:
        #print(f'Encoding one-hot for column: {column} \r')
        data[column] = LabelEncoder().fit_transform(y=data[column].fillna('0'))
```

```python
fig, ax = plt.subplots(figsize=(12,12))
ax.matshow(data.corr())
plt.xticks(range(len(data.columns)), [column for column in data.columns], rotation=90)
plt.yticks(range(len(data.columns)), [column for column in data.columns])
plt.title(f'Prediction using: {nrows} samples', y=1.15)
plt.show()
```

### Note
Since there are too many columns to put in a single matrix, we need to break it into pieces.

```python
columns_to_lookat = ['Hobby', 'SkipMeals', 'WakeTime','HoursComputer','RaceEthnicity',
                     'OpenSource','Student','FormalEducation', 'UndergradMajor', 'YearsCoding','JobSatisfaction', 'CareerSatisfaction']

fig, ax = plt.subplots(figsize=(10,10))
ax.matshow(data[columns_to_lookat].corr())
plt.xticks(range(len(columns_to_lookat)), [column for column in data[columns_to_lookat].columns], rotation=45)
plt.yticks(range(len(columns_to_lookat)), [column for column in data[columns_to_lookat].columns], rotation=45)
plt.title(f'Prediction using: {nrows} samples', y=1.15)
plt.ylabel('Columns Y')
plt.xlabel('Columns X')
plt.margins(.1)
plt.title(f'Prediction using: {nrows}. Shows the correlation between different columns')
plt.show()
```
